```python
price=list(df["Price"])
```

```python
fig=plt.figure()
plt.hist(x= price,edgecolor = "black",bins =2000)
plt.title("Histogram of price" )
plt.xlabel("price")
#plt.xticks([100,200,300,400,500,600,800,900,100])
plt.ylabel("Number of houses")
plt.show()
```

```python
rooms_count = df["Rooms"].value_counts()
rooms_count = dict(rooms_count)

a = rooms_count.keys()
b= rooms_count.values()
fig=plt.figure(figsize=(8,6))
plt.pie(x=a,labels=a)
plt.legend()
plt.show()
```

```python
region_count=df["Regionname"].value_counts()
region_count=dict(region_count)
r=list(region_count.keys())
rc=list(region_count.values())
```

```python
fig= plt.figure(figsize=(12,9))
plt.pie(x=rc,labels=r)
plt.legend()
plt.show()
```

```python
type_count=df["Type"].value_counts()
type_count=dict(type_count)
k=list(type_count.keys())
s=list(type_count.values())
fig=plt.figure(figsize=(12,10))
plt.title("TYPES OF HOUSES")
plt.xlabel("Houses")
plt.ylabel('No.of.Houses')
plt.bar(x=k,height=s,edgecolor = "red",color='black')
plt.show()
```

```python
price=df["Price"]
price=list(price)
distance=df["Distance"]
distance=list(distance)
fig=plt.figure(figsize=(10,8))
plt.plot(distance, price, marker=".",linestyle="",color='pink',markeredgecolor='black')
plt.show()
```
